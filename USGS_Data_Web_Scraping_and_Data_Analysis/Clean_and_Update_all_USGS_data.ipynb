```python
import glob
import pandas as pd
import numpy as np
import os.path
```

```python
pd.read_csv('https://raw.githubusercontent.com/sunnypark1026/sunnypark/master/USGS/usgs_state_1957.csv').head()
```

```python
# We have usgs_state_1957 ~ usgs_state_2020
# Need to add 'Year' Column 
# Need to remove 'Unnamed' columns

for i in range(1957,2021):
    k = pd.read_csv('https://raw.githubusercontent.com/sunnypark1026/sunnypark/master/USGS/usgs_state_{0}.csv'.format(i))
    # add 'Year' Column 
    k['Year'] = i
    # remove 'Unnamed' columns
    k = k[['State','Year','State Debt','Local Debt','State and Local Debt',
           'Gross State Product','Real State Growth %', 'Population (million)']] 
    k.to_csv('state_{0}new.csv'.format(i))
```

```python
# Use glob 
glob.glob('*new.csv')
```

```python
allfile_list = glob.glob(os.path.join('/Users/sunheepark/MyProject/USGS/Newly_Created_USGS_files','*new.csv'))
```

```python
# create an empty list
alldata = []
for file in allfile_list:
    df = pd.read_csv(file)
    alldata.append(df)
```

```python
# Use concat function to combine whole dataset
dataCombine = pd.concat(alldata, axis=0, ignore_index=True)
dataCombine = dataCombine.sort_values(by ='Year')
dataCombine = dataCombine.reset_index()[['State','Year','State Debt','Local Debt','State and Local Debt',
           'Gross State Product','Real State Growth %', 'Population (million)']]
dataCombine
```
