__Keywords__:
Social media Influencers is a user who has established credibility in a specific industry, has access to a huge audience and can persuade others to act based on their recommendations. 
It is important to note that these individuals are not merely marketing tools, but rather social relationship assets with which brands can collaborate to achieve their marketing objectives.

__Caption__: A user can add a caption to an image to explain and contextualize the subject of the photo. Instagram captions can include hashtags and tags.


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/sunnypark1026/sunnypark/master/Data_Analysis/Instagram-Influencer_Project/all_data.csv')
df = df[['Username', 'Category', 'Followers', 'Followees','Posts']]
# get category as 'fashion' only
df = df[df.Category == 'fashion']
# influencers who have more than 6,000 followers
# https://www.petalcard.com/blog/money-instagram-influencer-make
df = df[df.Followers >= 6000]
df.reset_index(drop=True, inplace=True)
df
```

```python
huge_if = df[df.Followers >= 100000]
huge_if
```

```python
mid_if = df[(df.Followers >= 10000) & (df.Followers < 100000)]
mid_if
```

```python
micro_if = df[df.Followers <10000]
micro_if
```

```python
huge_if['Level'] = 'Huge'
mid_if['Level'] = 'Mid'
micro_if['Level'] = 'Micro'
```
